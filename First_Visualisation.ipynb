```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import random
import os
```

```python

i = str(random.randint(1,292))
i = (3-len(i))*'0'+i

vol_path = fr"C:\Users\Moakher\Desktop\copy-of-pscc-data-challenge\train\volume\LUNG1-{i}_vol.nii"  
vol_contents = os.listdir(vol_path)
image_vol_path = os.path.join(vol_path, vol_contents[0])



seg_path = fr"C:\Users\Moakher\Desktop\copy-of-pscc-data-challenge\train\seg\LUNG1-{i}_seg.nii"
seg_contents = os.listdir(seg_path)
image_seg_path = os.path.join(seg_path, seg_contents[0])


lungseg_path = fr"C:\Users\Moakher\Desktop\copy-of-pscc-data-challenge\train\lungs_seg\LUNG1-{i}_lungseg.nii"
lungseg_contents = os.listdir(lungseg_path)
image_lungseg_path = os.path.join(lungseg_path, lungseg_contents[0])

nii  = nib.load(image_vol_path)
header = nii.header
## if(voxel_dims != (0.9765625, 0.9765625, 3.0)):
test_image_vol = nib.load(image_vol_path).get_fdata()
test_image_seg= nib.load(image_seg_path).get_fdata()
test_image_lungseg= nib.load(image_lungseg_path).get_fdata()
np.unique(test_image_seg)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=6, figsize=(20, 4), dpi=100)
slices_indices = [0,20,40,60,80,100]
for ind in range(6):
    slice = test_image_vol[:,:,slices_indices[ind]]
    axes[ind].imshow(slice,cmap='gray')
    axes[ind].axis('off')
plt.show()
```
